# Exploratory Analysis of Fraud dataset

The purpose of this notebook is to get an understanidng of the data in the dataset as well as make visualizations and identify which would be most important for detecting fraudulent transactions

```python
# Importing necessary modules
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

# Allows for matplotlib to show plots in the notebook
%matplotlib inline
```

```python
# Load the data
fraud_df = pd.read_csv("dataset/bank_transactions_data_2.csv")
fraud_df
```

```python
# Lets histogram some of these data
# For presentation purposes, lets make a array of plots with caragorical variables
fig, axes = plt.subplots(2,4, figsize=(17,8))
sns.countplot(y="TransactionType", data=fraud_df, ax=axes[0,0])
fraud_df["Channel"].hist(ax=axes[0,1])
sns.countplot(x="CustomerOccupation", data=fraud_df, ax=axes[0,2])
sns.countplot(x="MerchantID", data=fraud_df, ax=axes[0,3])
sns.countplot(x="LoginAttempts", data=fraud_df, ax=axes[1,0])
# fraud_df["Channel"].hist(ax=axes[1])

plt.show()
```

```python
# There seems to be a lot of categorical data, 
sns.pairplot(fraud_df)
plt.show()
```
